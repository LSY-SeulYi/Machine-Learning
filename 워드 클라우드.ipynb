```python
import konlpy
from konlpy.tag import Okt
from collections import Counter
import csv
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from pygments.lexers.csound import newline

filename = "conmpanyname1.txt"
f = open(filename, 'r', encoding='utf-8')
jobs = f.read()

# Okt 객체 생성
okt = Okt()
noun = okt.nouns(jobs)
for i,v in enumerate(noun):
    if len(v)<2:
        noun.pop(i)
count = Counter(noun)
word = dict(count.most_common())
word
print(word)
```

```python
word = {key: value for key, value in word.items() if value >=2}
```

```python
wordcloud = WordCloud(font_path = 'NanumGothic-Regular.ttf', background_color = 'white', colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(word)

plt.figure(figsize = (15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
